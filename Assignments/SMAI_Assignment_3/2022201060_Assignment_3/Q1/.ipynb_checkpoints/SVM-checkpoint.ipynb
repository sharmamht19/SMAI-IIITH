```python
# Helper function to visualize 2D points X with labels +1 or -1
def plot_points(X, Y):
    reds = []
    blues = []
    for i in range(Y.shape[0]):
        if Y[i] == 1:
            reds.append(X[i])
        else:
            blues.append(X[i])
    reds = np.array(reds)
    blues = np.array(blues)
    plt.scatter(reds[:, 0], reds[:, 1], c = 'r')
    plt.scatter(blues[:, 0], blues[:, 1], c = 'b')
```

In the next cell you are required to find the Quadratic Constrained Quadratic form of SVM **without soft constraints** from the basic principles (check the svm slides uploaded in the portal). Start with the idea that SVM tries to maximize the margin and then derive the form which you can feed to the solver.

Please print the values of w, b. 

**Note: You are also required to upload a page of the derivation. You may also type it in markdown here**
![derivation](./formulation.jpeg)
**Note: You cannot use sklearn SVM functions here**

```python
def solve_SVM(X,Y):
 #weight Matrix.
  n,dim=X.shape
  w=cp.Variable(dim)
  #gama term.
  b=cp.Variable()
  # objective function - (1/2)*(||W||^2).
  obj=0.5*cp.square(cp.norm(w))
  obj_min=cp.Minimize(obj)
  constraints=[ Y[i] * (w.T @ X[i] + b)-1 >= 0 for i in range(n)]
  p=cp.Problem(obj_min,constraints)
  p.solve()
  return p.status,w.value,b.value
```

```python
status,w,b=solve_SVM(X,y)
print("Optimal status and values :")
print(status)
print("b : ",b)
print("w : ",w)
```

In the next cell you are required to visualize the boundary predicted by the solver. You may want to revise up on how to plot the a line given w, b. Please plot w vector as well as the margin lines.

Create a function to plot the line of separation as you would require this frequently in the subsequent parts of the assignment.

```python
def visualize(w,b,X,Y,min_x,max_x):
  y_offset = -w[0] / w[1]
  # make the x-axis space for the data points
  X_space = np.linspace(min_x, max_x)
  y_space = y_offset * X_space - b / w[1]
  y_space_sv1=y_offset * X_space - b / w[1] +1/w[1]
  y_space_sv2=y_offset * X_space - b / w[1] -1/w[1]
  # plot the decision boundary
  plt.figure(figsize=(8, 8))
  plt.plot(X_space, y_space, '--',label='Decision boundary')
  plt.plot(X_space, y_space_sv1, 'k-',label='Hyperplane')
  plt.plot(X_space, y_space_sv2, 'k-')

  # show the plot visually
  plt.scatter(X[:, 0], X[:, 1], c=Y)
  plt.legend()
  plt.show()
```

```python
visualize(w,b,X,y,-4,4)
```

In the next cell you are required to modify your framing of optimisation problem to incorporate soft constraints also known as slack. Plot the decision boundary, w and margin lines for various values of C: $0.01$, $0.1$, $3$, $10$, $100$.

Explain your results theoretically. How does the boundary change with varying values of C? Why?

**Note: Use the modified X and y i.e after adding points in previous cell**

```python
def solve_SVM_with_slack(X,Y,c):
 #weight Matrix.
  m,dim=X.shape
  w=cp.Variable(dim)
  #slack variable
  s=cp.Variable(m)
  #gama term.
  b=cp.Variable()
  # objective function - (1/2)*(||W||^2+ sum(mistaks)).
  obj=0.5*cp.square(cp.norm(w))
  obj_min=cp.Minimize(obj+c*cp.sum(s))
  constraints=[]
  for i in range(m):
    constraints.append(Y[i] * (w.T @ X[i] + b)-1 + s[i]>= 0)
    constraints.append(s[i]>=0)
  p=cp.Problem(obj_min,constraints)
  p.solve()
  return p.status,w.value,b.value
```

###Explaination : 
Here in the SVM, the parameter C controls the balance between maximizing the margin between the classes and minimizing the classification error. A lower value of C prioritizes a wider margin over correct classification of all examples, resulting in a more general boundary that may include some misclassified examples just like we can see when value of c is 0.01 and 0.1 some misclassified points are inside the boundary. Conversely, a higher value of C prioritizes correct classification of all examples over the margin width, resulting in a more specific boundary that may exclude some misclassified examples just like here for higher value of c as 3, 10 and 100.

Take the case of C = 100 and remove the support vectors. Does the boundary change after the deletion? Explain why?

```python
def remove_sup_vect(w,b,X,Y,c):
  #weight Matrix.
  m,dim=X.shape
  #slack variable
  x_index=[]
  for i in range(m):
    # print(int(Y[i] * (X[i]@w.T  + b) -1))
    if(abs(Y[i] * (X[i]@w.T  + b) -1 )<0.00001):
      x_index.append(i)
    
  X = np.delete(X, x_index, axis=0)
  Y = np.delete(Y, x_index)
  return X,Y
```

```python
X_updated,y_updated=remove_sup_vect(w,b,X_new,y_new,100)
print(X_new.shape)
print(X_updated.shape)
print(X_updated)
print(y_updated)
```

```python
status,w,b=solve_SVM_with_slack(X_updated,y_updated,100)
visualize(w,b,X_updated,y_updated,-4,4)
```

###Explaination : 
Yes, As we can see when we are removing  support vectors that lies directly on the decision boundary (as dist will be nealy 0 from decision boundary), the boundary  changing because that support vector plays a key role in defining the boundary. Also the decision boundary moves closer to the remaining support vectors.<br>
  So, whether or not the decision boundary changes after removing support vectors depends on which support vectors we remove and how close they are to the boundary. If we remove support vectors far from the boundary, the boundary is unlikely to change. However, if we remove support vectors close to or on the boundary, the boundary is likely to shift.

```python
# Import Relevant Packages
from sklearn.svm import SVC
import numpy as np
from mnist.loader import MNIST
from tabulate import tabulate
```

```python
mndata = MNIST('samples')

# Load training dataset
images, labels = mndata.load_training()
l = len(labels)
images_train = np.array(images)
labels_train = np.array(labels)

print(images_train.shape)
print(labels_train.shape)
```

```python
# Load testing dataset
images, labels = mndata.load_testing()
l = len(labels)
images_test = np.array(images)
labels_test = np.array(labels)

print(images_test.shape)
print(labels_test.shape)
```

- Tweak different parameters like the C Parameter and gamma parameter of the Linear SVM and report the results.
- Experiment different kernels (Linear, Polynomial, RBF, Sigmoid) for classification and report the results.

Report accuracy score, F1-score, Confusion matrix and any other metrics you feel useful.

```python
# use this subset of data for training and testing

batch = 3000
X_train = images_train[:batch,:]
y_train = labels_train[:batch]

X_test = images_test[batch:2*batch,:]
y_test = labels_test[batch:2*batch]
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report

def get_y_Pred_SVM(method,c_value,g):
  svm =SVC(kernel=method,C=c_value,gamma=g)
  svm.fit(X_train, y_train)
  y_pred=svm.predict(X_test)
  return y_pred
c_values=[0.1,1,5,10,100]
gamma_values=[0.1,1,5,10,100]
```

```python
scores=[]
scores.append(["C","gamma","Accuracy"])
for c in c_values:
  for g in gamma_values:
    y_pred=get_y_Pred_SVM("linear",c,g)
    scores.append([c,g,accuracy_score(y_test, y_pred)])
```

```python

print(tabulate(scores, headers='firstrow', tablefmt='fancy_grid'))
```

```python
methods=["linear","poly", "rbf", "sigmoid"]
entries=[]
confusion_matrices=[]
reports=[]
entries.append(["Method","F1 score","Accuracy"])
for method in methods:
  y_pred=get_y_Pred_SVM(method,1,1)
  f1 = f1_score(y_test, y_pred, average='weighted')
  acc=accuracy_score(y_test, y_pred)
  c_m = confusion_matrix(y_test, y_pred)
  report = classification_report(y_test, y_pred)
  
  entries.append([method,f1,acc])
  confusion_matrices.append(c_m)
  reports.append(report)
  

print(tabulate(entries, headers='firstrow', tablefmt='fancy_grid'))
```
